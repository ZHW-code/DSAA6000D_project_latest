## 3. Add node(companies) and edges(relationships) to knowledg graph

```python
# Create the nodes of companies
for index, row in companies.iterrows():
    node = Node(row['company_name'], name=row['company_name'], ID = row[':ID'])
    graph.create(node)
```

```python
# Create the relation edges between companies
#COMPETE
for index, row in compete.iterrows():
    start_node = graph.nodes.match(companies.loc[companies[':ID'] == row[':START_ID'], 'company_name'].values[0], ID=row[':START_ID']).first()
    end_node = graph.nodes.match(companies.loc[companies[':ID'] == row[':END_ID'], 'company_name'].values[0], ID=row[':END_ID']).first()
    compete_relationship = Relationship(start_node, row[':TYPE'], end_node, name=row[':TYPE'])
    graph.create(compete_relationship)
    
for index, row in compete.iterrows():
    start_node = graph.nodes.match(companies.loc[companies[':ID'] == row[':END_ID'], 'company_name'].values[0], ID=row[':END_ID']).first()
    end_node = graph.nodes.match(companies.loc[companies[':ID'] == row[':START_ID'], 'company_name'].values[0], ID=row[':START_ID']).first()
    compete_relationship = Relationship(end_node, row[':TYPE'], start_node, name=row[':TYPE'])
    graph.create(compete_relationship)
```

## 4. Export the knowledge graph to a .csv file and make the label of the relationships as the GNN input data

```python
import csv
import networkx as nx
from neo4j import GraphDatabase
driver = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "20001028zhw"))

# 将Neo4j知识图谱数据保存为CSV文件
def export_graph_to_csv():
    with driver.session() as session:
        result = session.run("MATCH (n)-[r]->(m) RETURN n, r, m")
        with open("D:/HKUST(GZ)/Graph analysis/project/knowledge_graph.csv", "w", newline = "") as file:
            writer = csv.writer(file)
            writer.writerow(["start_node_name", "start_node_id", "end_node_name", "end_node_id", "relationship_type"])
            for record in result:
                start_node = record["n"]
                end_node = record["m"]
                relationship = record["r"]
                start_node_name = list(start_node.labels)[0]
                end_node_name = list(end_node.labels)[0]
                start_node_id = start_node["ID"]  # 根据实际情况提取节点属性
                end_node_id = end_node["ID"]  # 根据实际情况提取节点属性
                relationship_type = relationship.type
                writer.writerow([start_node_name, start_node_id, end_node_name, end_node_id, relationship_type])
```

```python
def read_graph_from_csv():
    graph = nx.Graph()
    with open("D:/HKUST(GZ)/Graph analysis/project/knowledge_graph.csv", "r") as file:
        reader = csv.reader(file)
        next(reader)  # 跳过标题行
        for row in reader:
            start_node_name, start_node_id, end_node_name, end_node_id, relationship_type = row
            graph.add_node(start_node_name, property=start_node_id)  # 根据实际情况添加节点属性
            graph.add_node(end_node_name, property=end_node_id)  # 根据实际情况添加节点属性
            graph.add_edge(start_node_name, end_node_name, type=relationship_type)
    return graph
```

```python
# 调用函数进行保存和读取
export_graph_to_csv()
graph = read_graph_from_csv()
print(graph.nodes)
print(graph.edges)
```
